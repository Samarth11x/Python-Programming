# OpenCV Basics


```python
from pathlib import Path
import cv2
import sys

script_dir = Path.cwd()
img_path = script_dir / "image.jpg"

img = cv2.imread(str(img_path))
if img is None:
    print(f"Error: could not load {img_path}")
    sys.exit(1)

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print("Image loaded successfully!")
```

## Resolution

```python
height, width, channels = img.shape
print(f"Resolution: {width} x {height}")
```

## Channels

```python
print(f"Channels: {channels}")
```

## Shape

```python
print(img.shape)
```

## RGB Matrix

```python
print(img_rgb)
```

## Split RGB Channels

```python
b,g,r=cv2.split(img_rgb)
print("Blue Channel:\n",b)
print("Green Channel:\n",g)
print("Red Channel:\n",r)
```

## imread()

```python
img2=cv2.imread(str(img_path))
print(img2.shape)
```

## imwrite()

```python
cv2.imwrite("output_image.jpg", img)
print("Saved as output_image.jpg")
```

## imshow()

```python
cv2.imshow("Original Image", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Show RGB Channels

```python
cv2.imshow("RGB Image", img_rgb)
cv2.imshow("Blue Channel", b)
cv2.imshow("Green Channel", g)
cv2.imshow("Red Channel", r)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## RGB to BGR

```python
img_bgr=cv2.cvtColor(img_rgb,cv2.COLOR_RGB2BGR)
cv2.imshow("BGR",img_bgr)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Grayscale

```python
cv2.imshow("Gray",gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Color Comparison

```python
cv2.imshow("BGR",img)
cv2.imshow("RGB",img_rgb)
cv2.imshow("Gray",gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Function to call a image

```python
def display(winname , img):
    cv2.imshow(winname, img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

# Display img

```python
display("winname", img)
```

# Access acc to Rows and Col

```python
img.shape
```

```python
display("winname", img)
```

```python
Corners = img[0:100, 0:100] 
display("Corners", Corners)
```

# Change 1st 100 rows and col to green color

```python
Green = (0, 255, 0 )
img[0:100, 0:100] = Green
display('manipulate',img)
```

# Crop using rows and col ranges

```python
crop = img[100:300, 150:400]

cv2.imshow("Crop", crop)
cv2.waitKey(0)  
cv2.destroyAllWindows()
```

# Resize Image 

```python
resized = cv2.resize(img, (300, 200))
smaller = cv2.resize(img, None, fx=0.5, fy=0.5)
cv2.imshow('resized', resized)
cv2.waitKey(0)  
cv2.destroyAllWindows()
```

# Flip Image

```python
horizontal_flip = cv2.flip(img, 1)
vertical_flip = cv2.flip(img, 0)
both_flip = cv2.flip(img, -1)

cv2.imshow("Horizontal Flip", horizontal_flip)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imshow("Vertical Flip", vertical_flip)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imshow("Both Flip", both_flip)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Summary
- imread(): Reads image
- imwrite(): Saves image
- imshow(): Displays image
- cvtColor(): Converts color spaces
- split(): Separates color channels
- shape: Gives height, width and channels.
- display() : function used to display image
- img.shape : used to manipulate the image acc to rows and columns 
- crop : crop using rows and column ranges
- resize : Used to resize the image based on scale
- flip : USed to flip the image Horizontally and Vertically
