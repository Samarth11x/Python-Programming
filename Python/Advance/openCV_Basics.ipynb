# OpenCV Basics


```python
from pathlib import Path
import cv2
import sys

script_dir = Path.cwd()
img_path = script_dir / "image.jpg"

img = cv2.imread(str(img_path))
if img is None:
    print(f"Error: could not load {img_path}")
    sys.exit(1)

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print("Image loaded successfully!")
```

## Resolution

```python
height, width, channels = img.shape
print(f"Resolution: {width} x {height}")
```

## Channels

```python
print(f"Channels: {channels}")
```

## Shape

```python
print(img.shape)
```

## RGB Matrix

```python
print(img_rgb)
```

## Split RGB Channels

```python
b,g,r=cv2.split(img_rgb)
print("Blue Channel:\n",b)
print("Green Channel:\n",g)
print("Red Channel:\n",r)
```

## imread()

```python
img2=cv2.imread(str(img_path))
print(img2.shape)
```

## imwrite()

```python
cv2.imwrite("output_image.jpg", img)
print("Saved as output_image.jpg")
```

## imshow()

```python
cv2.imshow("Original Image", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Show RGB Channels

```python
cv2.imshow("RGB Image", img_rgb)
cv2.imshow("Blue Channel", b)
cv2.imshow("Green Channel", g)
cv2.imshow("Red Channel", r)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## RGB to BGR

```python
img_bgr=cv2.cvtColor(img_rgb,cv2.COLOR_RGB2BGR)
cv2.imshow("BGR",img_bgr)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Grayscale

```python
cv2.imshow("Gray",gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Color Comparison

```python
cv2.imshow("BGR",img)
cv2.imshow("RGB",img_rgb)
cv2.imshow("Gray",gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Summary
- imread(): Reads image
- imwrite(): Saves image
- imshow(): Displays image
- cvtColor(): Converts color spaces
- split(): Separates color channels
- shape: Gives height, width and channels.
